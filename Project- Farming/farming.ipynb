## Applying Decision Tree Algorithm 

```python
# Decision Tree Classifier
```

```python
import pandas as pd
from sklearn.metrics import classification_report,precision_score,recall_score,f1_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#### Load the dataset :

```python
df = pd.read_csv("Farming_cleaned.csv")
```

```python
df
```

```python
df.dropna(inplace=True)

x= df.drop('label',axis=1)
y= df[['label']]
```

```python
x
```

```python
y
```

#### Split the dataset into training and testing sets :

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

#### Instantiate the Decision Tree Classifier :


```python
# Making an object dt
dt= DecisionTreeClassifier()
```

#### Train the model :

```python
dt.fit(x_train,y_train)
```

```python
# Finding score of testing data

dt.score(x_test,y_test)
```

```python
# Finding score of training data

dt.score(x_train,y_train)
```

#### Make predictions on the test set :

```python
pred = dt.predict(x_test)
```

```python
pred=pd.DataFrame(pred)
```

```python
pred
```

### Classification Report:-

```python
print(classification_report(y_test,pred))
```

```python
### Analyzing the precision score, recall score and F1 score of the algorithm


# print(precision_score(y_test,pred))
# print(recall_score(y_test,pred))
# print(f1_score(y_test,pred))
```

#### Finding the confusion matrix :

```python
print(confusion_matrix(y_test,pred))
```

### Evaluate the model :-

```python
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.2f}")
```

## Applying Random Forest Algorithm

```python
# Random Forest Classifier
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,precision_score,recall_score,f1_score, confusion_matrix
```

#### Load the dataset :

```python
df = pd.read_csv("Farming_cleaned.csv")
```

```python
df
```

```python
x= df.drop('label',axis=1)
y= df[['label']]
```

```python
x
```

```python
y
```

#### Split the dataset into training and testing sets :

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

#### Instantiate the Random Forest Classifier :

```python
# n_estimators: The number of trees in the forest.
# random_state: Controls the randomness of the bootstrapping of the samples used when building trees.

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
```

#### Train the model :


```python
rfc.fit(x_train, y_train)
```

```python
# Finding score of testing data

rfc.score(x_test,y_test)
```

```python
# Finding score of training data

rfc.score(x_train,y_train)
```

#### Make predictions on the test set :

```python

y_pred = rfc.predict(x_test)
```

```python
y_pred
```

### Evaluate the model :-


```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Random Forest Classifier: {accuracy:.2f}")
```
